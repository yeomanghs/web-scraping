```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from datetime import datetime

driver_path = 'D:/Users/figohjs/Desktop/Alert/Source/chromedriver.exe'
resultList = []
website = "https://www.outbreak.my/map"
ignoredExceptions=(NoSuchElementException,StaleElementReferenceException,)
options = webdriver.ChromeOptions()
start =datetime.now()
# options.add_argument("start-maximized")
options.add_argument('headless')    
driver = webdriver.Chrome(executable_path = driver_path, options = options)
driver.get(website)
# driver.set_page_load_timeout(30)

#find total number of markers 
className = 'leaflet-marker-icon leaflet-div-icon leaflet-zoom-animated leaflet-interactive'
xPath = '//div[@class="%s"]'%className
element = WebDriverWait(driver, 20, ignored_exceptions = ignoredExceptions)\
                        .until(expected_conditions.presence_of_all_elements_located((By.XPATH, xPath)))
totalElement = len(element)
noAttempt = 2

def copyPopUp(num):     
    #click marker
    className = 'leaflet-marker-icon leaflet-div-icon leaflet-zoom-animated leaflet-interactive'
    xPath1 = '//div[@class="%s"][%s]'%(className, num)
    element = WebDriverWait(driver, 20,ignored_exceptions = ignoredExceptions)\
                            .until(expected_conditions.presence_of_element_located((By.XPATH, xPath1)))
    driver.execute_script("arguments[0].click();", element)

    #extract popup content
    className = "leaflet-popup-content"
    xPath2 = '//div[@class="%s"]'%className
    ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
    element = WebDriverWait(driver, 20, ignored_exceptions = ignoredExceptions)\
                            .until(expected_conditions.presence_of_element_located((By.XPATH, xPath2)))
    a = element.find_element_by_xpath(xPath2)
    return a.text


for i in range(1, totalElement + 1):
    driver.quit()
    driver = webdriver.Chrome(executable_path = driver_path, options = options)
    driver.get(website)
    currentText = copyPopUp(i)
    #         element = driver.find_element_by_xpath(xPath1)
    #         driver.execute_script("arguments[0].click();", element)

    # #click marker
    # className = 'leaflet-marker-icon leaflet-div-icon leaflet-zoom-animated leaflet-interactive'
    # xPath1 = '//div[@class="%s"][1]'%className
    # element = driver.find_element_by_xpath(xPath1)
    # element = element.find_element_by_xpath(xPath1)
    # driver.execute_script("arguments[0].click();", element)
#     if currentText == '':
#         for j in range(1, noAttempt+1):
#             currentText = copyPopUp(i)
#             if currentText != '':
#                 break
        
    resultList.append(currentText)    
    driver.quit()
    
end = datetime.now()
```

```python
#Original data

import csv
#remove non-ASCII character
resultList2 = [''.join([j for j in i if ord(j) <= 127]) for i in resultList]

filename = 'D:/Users/figohjs/Desktop/Alert/Data/' + str(datetime.now().date()) + '_Covid19_OriData.csv'
with open(filename, 'w', encoding = 'utf-8', newline = '\n') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    for row in resultList2:
        wr.writerow([row])
```

```python
#Processed data
newsList = []
processedResultList = []
for news in resultList2:
    #replace duplicates with ''
    if news not in newsList:
        #dun want empty string
        if news != '':
            newsList.append(news)
            processedResultList.append(news)
    else:
         processedResultList.append('')   
```

```python
#213 out of 251 markers
filename = 'D:/Users/figohjs/Desktop/Alert/Data/' + str(datetime.now().date()) + '_Covid19_ProcessedData.csv'
with open(filename, 'w', encoding = 'utf-8', newline = '\n') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    for row in processedResultList:
        wr.writerow([row])
```
