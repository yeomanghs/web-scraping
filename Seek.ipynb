Q1. Write a function in python to sum up a given set of numbers other than itself Input: An array of n integers nums, 
    Output: An array output such that output[i] is equal to the sum of all the elements of nums except nums[i]. 
    For example, given [1,2,3,4], return [9,8,7,6].

```python
import numpy as np
#input
input_list = np.array([1,2,3,4])
input_list
```

```python
#function
def GetSum(array):
    return np.array([sum([num for num in array if list(array).index(num)!=i]) for i,j in enumerate(array)])
```

```python
#output
output_list = GetSum(input_list)
output_list
```

```python
def clean_string(x):
    #split by \r
    if type(x) == str:
        if re.search(', (\d+)',x.split('\r')[0]) and len(re.search(', (\d+)',x.split('\r')[0]).group(1)) == 4:
            pattern = re.compile('([%s]+ )(\d{2}), (\d{4})'%'|'.join(month_list))
            try:
                return re.search(pattern,x.split('\r')[0]).group(2) + ' ' + re.search(pattern,x.split('\r')[0]).group(1)\
                    + re.search(pattern,x.split('\r')[0]).group(3)
            except:
                return x
        elif re.search(', (\d+)',x.split('\r')[0]) and len(re.search(', (\d+)',x.split('\r')[0]).group(1)) > 2:
            pattern = re.compile('((\d{2} )*[%s]+ \d+)'%'|'.join(month_list))
            try:
                return re.search(pattern,x).group(1)
            except:
                return x
        elif re.search('\d{2} [%s]+ (\d+)'%'|'.join(month_list),x) and len(re.search('\d{2} [%s]+ (\d+)'%'|'.join(month_list),x).group(1)) == 2:
            try:
                return re.search('(\d{2} [%s]+ \d+)'%'|'.join(month_list),x).group(1)
            except:
                return x
        else:
            return ''
    else:
        return ''
```

e) Clean text and generate new text from Job Responsibilities column: The new text shall not contain any stop words, 
    and the plural words shall be converted into singular words.

```python
import gensim
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
from nltk.stem import WordNetLemmatizer
#download if first time
#nltk.download('wordnet')wnl.lemmatize(i)
stopword_set = set(stopwords.words('english'))
tokenizer = RegexpTokenizer('\w+')
def clearStopwords(x):
    wnl = WordNetLemmatizer()
    if type(x) == str:
        tokenized_list = tokenizer.tokenize(x)
        cleantokenized_set = set(tokenized_list).difference(stopword_set)
        return ' '.join([wnl.lemmatize(i) for i in cleantokenized_set])
    else:
        return ''
    
df_subset['clean_JobRequirment'] = df_subset['JobRequirment'].map(clearStopwords)
```

Q3. Semantic similarity

a) Using the output from Q2, identify semantically similar job ads and group them together.
Job Description

```python
import gensim
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
#first time
#nltk.download('stopwords')
```

```python
#get tokenizer and stopwords set
tokenizer = RegexpTokenizer('\w+')
stopword_set = set(stopwords.words('english'))
```

```python
#tokenize panda series and remove stop words
cleantokenized_list = []
for i in df_subset['JobDescription']:
    if type(i) == str:
        tokenized_list = tokenizer.tokenize(i)
        cleantokenized_set = set(tokenized_list).difference(stopword_set)
        cleantokenized_list.append(cleantokenized_set)
    else:
        cleantokenized_list.append('')
#tag every sentence
result_list = []
for idx, doc in enumerate(cleantokenized_list):
    if doc!='':
        result_list.append(gensim.models.doc2vec.TaggedDocument(doc,[idx]))
```

```python
model = gensim.models.Doc2Vec(vector_size=300, min_count=0, alpha=0.025, min_alpha=0.025)
model.build_vocab(result_list)
```

```python
result_list
```

```python
#training of model
for epoch in range(10):
    print ('iteration'+str(epoch+1))
    model.epochs = epoch
    model.train(result_list,total_examples=model.corpus_count,epochs = model.epochs)
    model.alpha -= 0.002
    model.min_alpha = model.alpha
#saving the created model
model.save('doc2vec.model')
print('model saved')
```

```python
#loading the model
d2v_model = gensim.models.doc2vec.Doc2Vec.load('doc2vec.model')

index_list = [i[1][0] for i in result_list]
sscore_dict = {}
for no, i in enumerate(index_list):
    #get top 1000 similar ads and choose ads with similarity score than is more than 0.5
    sscore_index_list = [i[0] for i in d2v_model.docvecs.most_similar(i,topn = 1000) if i[1]>0.5 and i[0] in index_list]\
                            + [i]
    if sscore_index_list:
        sscore_dict[no] = sscore_index_list
        print('Group: ' + str(no) + ', Number of ads: ' + str(len(sscore_index_list)))
        #remove similar ads from list
        for i in sscore_index_list:
            index_list.remove(i)    
```

```python
#print grouping result based on Job Description
print('Number of groups: ' + str(len(sscore_dict)))
```

```python
#print one group based on JobDescription
df_groupedjob = df_subset[df_subset['JobDescription'].index.isin(sscore_dict[2])]
df_groupedjob
```

b) Extract best-selling music artists from https://en.wikipedia.org/wiki/List_of_bestselling_music_artists

```python
#send request to the url
htmlPage = urllib.request.urlopen("https://en.wikipedia.org/wiki/List_of_best-selling_music_artists")
```

```python
#beautify html 
htmlContent = BS(htmlPage,'html.parser')
```

```python
#store best-selling music artists in list
artists_list = []
for j in htmlContent.find_all(scope = 'row'):
    for k in j.find_all('a'):
        artists_list.append(k.get_text())
```

```python
#filter unwanted elements
bestartist_list = [i for i in artists_list if i not in ['Record labels',
                                     'Genres',
                                     'Release\nformats',
                                     'Charts',
                                     'Television']]
bestartist_list
```

c) Find out how many best-selling music artists are from 1990s

```python
period_list = []
for i in htmlContent.find_all('tr'):
    for j in i.select('td:nth-of-type(2)'):
        if j:
            period_list.append(j.get_text())
```

```python
#import modules
import pandas as pd
import re

#form pandas df
df_bestSelling = pd.DataFrame({'artist':pd.Series(bestartist_list),
                  'period':pd.Series([i for i in period_list if '\n\n' not in i])})

#get starting year
df_bestSelling['startingYear'] = df_bestSelling['period'].map(lambda x:int(re.search('(\d{4})',x).group(1)))
```

```python
#filter 1990s (1990-1999)
df_bestSelling_1990s = df_bestSelling[(df_bestSelling['startingYear']>=1990) & (df_bestSelling['startingYear']<2000)].copy()
df_bestSelling_1990s
```

```python
print('Number of best-selling music artists from 1990s: %s'%str(df_bestSelling_1990s.shape[0]))
```
