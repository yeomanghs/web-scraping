```python
#950 calls per day

#venue ids
idNo1 = '4c5ba4192815c9281618b067'
idNo2 = '5004f1d0e4b0b7eb526730d0'
idNo3 = '4d25d24b38d0236ac8f6eeb6'

venues_list = []
for VENUE_ID in [idNo1, idNo2, idNo3]:

    #multiple location id
    url = 'https://api.foursquare.com/v2/venues/{}?v={}&client_id={}&client_secret={}'.format(
        VENUE_ID,
        VERSION,
        CLIENT_ID, 
        CLIENT_SECRET,)

    results = requests.get(url).json()
    
    #get one for checking
    searchResults = results['response']
    
    v = results['response']
    
    tempList = [
                v['venue']['id'],
                v['venue']['name'], 
                v['venue']['location']['lat'], 
                v['venue']['location']['lng'],  
                v['venue']['categories'][0]['name'],
                ','.join(v['venue']['location']['formattedAddress']),
                v['venue']['photos']['count'],
                v['venue']['likes']['count'],
                ]
    
    #check if col exists for direct extract
    colList = ['dislike', 'rating', 'seasonalHours']
    for col in colList:
        if col in v['venue'].keys():
            tempList.append(v['venue'][col])
        else:
            tempList.append('')
    
    #review
    if len(v['venue']['tips']['groups'])>0:
        reviewInfo = v['venue']['tips']['groups'][0]['items']
        reviewList = [i['text'] for i in reviewInfo]
        tempList += ['|'.join(reviewList)]
    else:
        tempList.append('')
        
    #price
    tempList.append(v['venue']['price']['message'])
    
    venues_list.append(tempList)

dfVenues = pd.DataFrame(venues_list)
dfVenues.columns = [
              'Venue Id',
              'Venue Name', 
              'Venue Latitude', 
              'Venue Longitude', 
              'Venue Category',
              'Venue Address',
              'Number of Photo',
              'Number of Likes',
              'Number of Dislikes',
              'Rating',
               'Seasonal Hours',
              'Review',
              'Price']
#process dislikes
dfVenues['Number of Dislikes'] = dfVenues['Number of Dislikes'].map(lambda x:0 if x == False else x)

#source
# dfVenues['Source'] = 'FourSquareAPI'

dfVenues
```

```python
#example:
venues_list2 = []
for result in placesDict['results']:
    tempList = [
                result['place_id'],
                result['name'],
                result['geometry']['location']['lat'],
                result['geometry']['location']['lng'],
                ]
    
    for col in ['rating', 'user_ratings_total', 'vicinity']:
        if col in result.keys():
            tempList.append(result[col])
        else:
            tempList.append('')
    
    #category
    categoryInfo = result['types']
    categoryInfo = ','.join([i for i in categoryInfo if i not in ['point_of_interest', 'establishment']])
    tempList.append(categoryInfo)
    
    venues_list2.append(tempList)
```
